```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load audio file in the player
audio_path = "0000.wav"
ipd.Audio(audio_path)
```

```python
# load audio file
signal, sr = librosa.load(audio_path)
print(signal.max())
print(signal.min())

# signal += abs(signal.min())
# signal = (signal/signal.max())*2 -1
# print(signal.shape)
# print(signal.max())
# print(signal.min())
```

```python
# plot waveform
plt.figure(figsize=(18, 8))

#plt.plot(signal)
librosa.display.waveplot(signal, sr=sr, alpha=0.5)
plt.show()
```

```python
# derive spectrum using FT
ft = sp.fft.fft(signal)
print(ft.shape)
half_len = int(ft.shape[0]/2)
magnitude = librosa.power_to_db(np.absolute(ft)**2)
frequency = np.linspace(0, sr, len(magnitude)) 
```

```python
# plot spectrum
plt.figure(figsize=(18, 8))
plt.plot(frequency[:], magnitude[:]) # magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```
